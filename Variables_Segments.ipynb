```python
segment_id = 2
segment_filename = f'segments/segment_{segment_id}.csv'
segment = pd.read_csv(segment_filename, dtype = {'timestamp_utc': 'object',
 'dissolved_oxygen_percent_saturation': 'float64',
 'dissolved_oxygen_uncorrected_mg_per_l': 'float64',
 'salinity_psu': 'float64',
 'sensor_depth_measured_m': 'float64',
 'temperature_degree_c' : 'float64',
 'qc_flag_dissolved_oxygen_percent_saturation': 'str',
 'qc_flag_dissolved_oxygen_uncorrected_mg_per_l': 'str',
 'qc_flag_salinity_psu': 'str',
 'qc_flag_sensor_depth_measured_m': 'str',
 'qc_flag_temperature_degree_c': 'str'} , )

display(metadata.iloc[segment_id])
display(segment)

```

```python
# split into different measurement variables
variables = ['dissolved_oxygen_percent_saturation',
             'dissolved_oxygen_uncorrected_mg_per_l',
             'salinity_psu',
             'sensor_depth_measured_m',
             'temperature_degree_c']

for variable in variables:
    os.makedirs(variable, exist_ok=True)
    qc_flag = 'qc_flag_' + variable    
    series = segment[['timestamp_utc', variable, qc_flag]].dropna()

    series_filename = os.path.join(variable, f'{segment_id}_{variable}.csv')
    series.to_csv(series_filename)
```
